# 7.Identify at which time of the day, people are ordering more pizzas (morning/evening/afternoon/night)


```python
def a(value):
    if value>=6 and value<12:
        return "moring_order"
    elif value>=12 and value<16:
        return "after_noon_order"
    elif value>=16 and value<20:
        return "evening_order"
    else:
        return "night_order"

orders["Day_Time"]= orders["hour"] .apply(a)
orders
```

```python
gb = orders.groupby("Day_Time").agg({"order_id" : "count"})
gb.sort_values("order_id", ascending = False)
gb.idxmax()
```

# the largest amount of pizza is ordering in after_noon_order
